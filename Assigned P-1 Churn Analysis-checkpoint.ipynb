```python
# removing missing values
dataset.dropna(inplace=True)
dataset.isnull().sum()
```

```python
# to remove the customer IDs
df = dataset.iloc[:,1:]
df
```

```python
# conerting the predictor variable in a binary numeric variable
df["Churn"].replace(to_replace="Yes",value=1, inplace=True)
df["Churn"].replace(to_replace="No",value=0, inplace =True)

# To convert all the catagorical variables into dummy variables
df_dummies = pd.get_dummies(df)
df_dummies.head()
```

```python
# relation of churn with the other variables
plt.figure(figsize=(15,6))
df_dummies.corr()["Churn"].sort_values(ascending=False).plot(kind="bar",color=["teal","indianred","plum"])
```
